```python

import sklearn
from sklearn import cluster
import folium

import pandas as pd
import numpy as np
import random
from threading import Thread

import matplotlib.pyplot as plt
import matplotlib.cm as cm
#import matplotlib.colors as mpl_colors

from IPython.display import clear_output, display, HTML
import sys  # This is necessary for printing updates within a code block, 
            #   via sys.stdout.flush()
import time # Use time.sleep(secs) to sleep a process if needed
%matplotlib inline

print("Current environment directory:" + sys.prefix)
print("System version: "+sys.version)

def inline_map(m, width=500, height=400):
    """Takes a folium instance and embed HTML."""
    m._build_map()
    srcdoc = m.HTML.replace('"', '&quot;')
    embed = HTML('<iframe srcdoc="{}" '
                 'style="width: {}px; height: {}px; '
                 'border: none"></iframe>'.format(srcdoc, width, height))
    return embed
```

```python
myDf = pd.read_csv('Dummy_MappingData.csv')
myDf.head()
```

```python
colormapper = cm.ScalarMappable(cmap='rainbow')
colorsNEM1 = colormapper.to_rgba(myDf['NEM1'])
colorsNEM1[0]*256
```

```python
width, height = 500, 400
myMap = folium.Map(location=[40, -122], zoom_start=5, width=width, height=height)

# Plotting approach:

# Loop over all regions
#   For each region, loop over cluster labels
#     For each cluster label, loop over rows
#       If the row is a core node, plot with a big circle
#       If the row is not a core node, plot with a small circle


for row in myDf.index[0:50]:
    myMap.circle_marker(location=[myDf.loc[row,'latitude'],myDf.loc[row,'longitude']], 
                popup = myDf.loc[row,'address'],
                line_color=None,  fill_color=colorsNEM1[row]*256)

clear_output()                                    
print('Generating map...')                
inline_map(myMap)
```
