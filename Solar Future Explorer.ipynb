```python
import pandas as pd
import numpy as np
data = pd.read_csv('yearly_city_data.csv', sep=',')
cities = pd.read_csv('City_Locations.csv', sep=',')



#data.dtypes
dates = pd.date_range('1/1/2015', periods=365)
year_usage = pd.DataFrame(data['Bakersfield Meadows'].reshape(365,24),index=dates,columns = np.arange(24))
year_usage = pd.DataFrame.transpose(year_usage)
mean_hourly_load = pd.DataFrame(year_usage.mean(axis=1),columns = ['usage'])
mean_hourly_load[:3]
#year_usage[:3]
```

```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline  

x = np.arange(24)
plt.plot(x, mean_hourly_load['usage'])
#plt.plot(x, year_usage['2015-01-09 00:00:00'])
plt.title('Hourly Usage')
plt.show()
```

```python
average_city_data = pd.DataFrame(np.random.randn(24, 64),columns = data.columns)
average_city_data[:3]
dates = pd.date_range('1/1/2015', periods=365)

for x in range(0, 64):
    
    year_usage = pd.DataFrame(data.loc[:,x].reshape(365,24),index=dates,columns = np.arange(24))
    year_usage = pd.DataFrame.transpose(year_usage)
    mean_hourly_load = pd.DataFrame(year_usage.mean(axis=1),columns = ['usage'])
    average_city_data.loc[:,x] = mean_hourly_load
#def bigcityaverage(data)

average_city_data.columns = cities.columns
#avg_city_data = bigcityaverage(averag_city_data)
average_city_data[:3]
```

```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline  

x = np.arange(24)
plt.plot(x, average_city_data)
#plt.plot(x, year_usage['2015-01-09 00:00:00'])
plt.title('Hourly Usage')
plt.show()
```
