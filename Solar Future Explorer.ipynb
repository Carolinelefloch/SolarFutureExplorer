```python
import pandas as pd
import numpy as np
data = pd.read_csv('Arcata_Usage.csv', sep=',')

#data.dtypes
dates = pd.date_range('1/1/2015', periods=365)
year_usage = pd.DataFrame(data['Electricity:Facility [kW](Hourly)'].reshape(365,24),index=dates,columns = np.arange(24))
year_usage = pd.DataFrame.transpose(year_usage)
mean_hourly_load = pd.DataFrame(year_usage.mean(axis=1),columns = ['usage'])
mean_hourly_load
year_usage
```

```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline  

x = np.arange(24)
plt.plot(x, mean_hourly_load['usage'])
#plt.plot(x, year_usage['2015-01-09 00:00:00'])
plt.title('Hourly Usage')
plt.show()
```
